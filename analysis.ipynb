```python
# which countries have launched from Cape Canaveral?

df_launch_site = df.select(['Country of Operator/Owner', 'Launch Site'])

df_launch_site = df_launch_site.filter(pl.col('Launch Site') == 'Cape Canaveral')

df_launch_site['Country of Operator/Owner'].value_counts(sort=True) # expect most to the US
```

```python
# where does turkey launch from the most?

df_turkey = df.select(['Country of Operator/Owner', 'Launch Site'])


df_turkey = df_turkey.filter(pl.col('Country of Operator/Owner') == 'Turkey')

df_turkey['Launch Site'].value_counts(sort=True) 

# Oh interesting, TR launches from Jiuquan, China? and Satish Dhawan Space Centre, India? and Dombarovskiy, Russia?
# Let's see what those missions are about.
```

```python
# search for Georgia Tech satellites

big_df = df.select(['Name of Satellite, Alternate Names', 'Country of Operator/Owner', 'Operator/Owner', 'Users', 'Purpose', 'Detailed Purpose', 'Class of Orbit', 'Type of Orbit', 'Date of Launch', 'Expected Lifetime (yrs.)', 'Contractor', 'Country of Contractor', 'Launch Site', 'Launch Vehicle',])

gt_df = big_df.filter(pl.col('Operator/Owner') == 'Georgia Institute of Technology')

gt_df

# We can see that Georgia Tech has launched 3 satellites. Go Jackets!
```

```python
# what about the University of Georgia?
uga_df = big_df.filter(pl.col('Operator/Owner').str.contains("Georgia")) # I'm not quite sure how they'd be listed in the database, so I'm just going to search for "Georgia" in the Operator/Owner column. 

uga_df
```

```python
# Nope, no satellites for UGA. I know they're working on a project called MEMESat, but it's not in the database given it's not launched yet.
# http://www.smallsat.uga.edu/missions
```

```python
# Does MIT have any satellites as a contractor?

list_of_possible_names_for_mit = ['Massachusetts Institute of Technology', 'MIT', 'Massachusetts Institute of Technology Lincoln Laboratory', 'Massachusetts Institute of Technology Lincoln Laboratory (MITLL)']

list_of_satellites_for_mit = []
for name in list_of_possible_names_for_mit:
    list_of_satellites_for_mit.append(big_df.filter(pl.col('Contractor').str.contains(name)))
```

```python
list_of_satellites_for_mit
```

```python
# convert the list of dataframes to a single dataframe

mit_df = pl.concat(list_of_satellites_for_mit)
```

```python
mit_df # MIT has launched 5 satellites as a contractor, but only operates two of them.
```
